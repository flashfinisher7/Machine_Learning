```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
```

```python
dataset=pd.read_csv(r"C:\Users\mvsan\Downloads\Market_Basket_Optimisation.csv",header=None)
dataset.head()
```

```python
dataset.shape
```

```python
#initializing the list of transcation that create empty list
transcations=[]
#going to first for loop for all transcations
for i in range(0,7501):
#     print(i)
    #j for other for loop 
    transcations.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
# transcations
type(transcations)
```

```python
#training apriori on dataset
#transcation is inputs and rules are outputs
from apyori import apriori 
#min_support=3*7/7500=0.003
rules = apriori(transcations,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2)
```

```python
#visualizing the data
results = list(rules) 
```

```python
print(results[0])  
```
